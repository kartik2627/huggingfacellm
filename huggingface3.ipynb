```python
from datasets import load_dataset

langs = ["en", "de", "fr", "hi", "zh"]
fracs = [1.0, 0.5, 0.3, 1.0, 0.7]  # Customize sampling fraction per language

panx_ch = {}

for lang, frac in zip(langs, fracs):
    print(f"\nLoading PAN-X.{lang} with fraction {frac}")

    # Load dataset
    ds = load_dataset("xtreme", name=f"PAN-X.{lang}")

    # Initialize nested dict for the language
    panx_ch[lang] = {}

    # Shuffle and sample each split
    for split in ds:
        num_samples = int(frac * ds[split].num_rows)
        panx_ch[lang][split] = ds[split].shuffle(seed=0).select(range(num_samples))
        print(f"  - {split}: {num_samples} samples")
```

```python
import pandas as pd

pd.DataFrame(
    {lang: [panx_ch[lang]["train"].num_rows] for lang in langs},
    index=["number of training examples"]
)
```

```python
tags = panx_ch["en"]["train"].features["ner_tags"].feature
print(tags)
```

```python
def create_tag_names(batch):
    return {"ner_tags_str": [tags.int2str(idx) for idx in batch["ner_tags"]]}

panx_dedup = {}
for split in panx_ch["de"]:
    panx_dedup[split] = panx_ch["de"][split].map(create_tag_names)
```

```python
de_example = panx_dedup["train"][0]
pd.DataFrame([de_example["tokens"], de_example["ner_tags_str"]], index=["Tokens", "Tags"])
```

```python
from collections import defaultdict, Counter
import pandas as pd
from datasets import load_dataset

# Load one language's dataset, for example: German
panx_de = load_dataset("xtreme", name="PAN-X.de")

# Get mapping from int tag to string
label_list = panx_de["train"].features["ner_tags"].feature.names

# Count B- tags (i.e., beginning of named entities)
split2freqs = defaultdict(Counter)

for split, dataset in panx_de.items():
    for row in dataset["ner_tags"]:
        for tag in row:
            tag_str = label_list[tag]
            if tag_str.startswith("B-"):
                tag_type = tag_str.split("-")[1]
                split2freqs[split][tag_type] += 1

# Convert to dataframe
df = pd.DataFrame.from_dict(split2freqs, orient="index").fillna(0).astype(int)
df
```
